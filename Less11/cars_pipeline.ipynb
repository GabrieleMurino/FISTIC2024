```python
# Import the neccessary modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.ensemble import RandomForestClassifier

import warnings
warnings.filterwarnings('ignore')
```

```python
path = 'cars_cleaned.csv'
df = pd.read_csv(path)
df
```

```python
df['brand'] = df['name'].str.split().str[0]
df = df.drop("name", axis=1)
df
```

```python
#df = pd.get_dummies(df, columns= ["fuel_type", "brand"], dtype=int, drop_first=True)
#df
```

```python
columns_to_ohe = ["fuel_type", "brand"]

OHE = ColumnTransformer([('ohe', OneHotEncoder(handle_unknown='ignore',drop='first'),columns_to_ohe )],
                        remainder='passthrough',  # Keep other columns unchanged
                        )

scaler = StandardScaler()
classifier = RandomForestClassifier(bootstrap=True,
                                    max_depth=80,
                                    max_features=2,
                                    min_samples_leaf=3,
                                    min_samples_split=8,
                                    n_estimators=100,
                                    )
#classifier = LogistiRegression()
```

```python
model_pipe = Pipeline([
                        ('Preprocessing features', OHE),
                        ('Scaling and standardize data', scaler),
                        ('Classifier', classifier)
                        ])
```
